```python
import numpy as np
from matplotlib import pyplot as plt
```

# Line Plot

```python
x = np.arange(1,11)
x
```

```python
y = 3*x
y
```

```python
plt.plot(x,y) # ploting x on x axis and y on y axix
plt.show()
```

```python
plt.plot([1,2,3,4,5],[2,3,4,0,6])
plt.show()
```

# Adding Title and Levels

```python
plt.plot(x,y,linestyle='--') # it will plot line in the graph
plt.title("This is title of Lien ploting")   # it will add the title in graph
plt.xlabel(" This is X axis")   # it will add a text below the x axis
plt.ylabel("This is Y axis")  # it will add a text below the y axis
plt.show()
```

# Changing Line Aesthetics

```python
plt.plot(x,y, color='r',linestyle='dashdot',linewidth=2) #color is g  means green fourth agrument represent the line style and last width
plt.show()
```

# Adding two line in the same plot

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
print(y1,y2)
```

```python
plt.plot(x, y1, color='m', linestyle = 'dotted', linewidth=2)
plt.plot(x, y2, color='b', linestyle = 'dashed', linewidth=2)
plt.title("Thsi is title")
plt.xlabel("This is x axis")
plt.ylabel("Thsi is y axis")
plt.grid(True)
plt.show()
```

# Adding sub-plots

```python
plt.subplot(1,3,1) # First arg represent the n-rows sec n-col thrid position of subplot
plt.title("Thsi is title")
plt.plot(x, y1, color='m', linestyle = 'dotted', linewidth=2)
plt.ylabel("Thsi is y axis")
plt.xlabel("This is x axis")


plt.subplot(1,3,3) # First arg represent the n-rows sec n-col thrid position of subplot
plt.title("This is first title")
plt.plot(x, y2, color='b', linestyle = 'dashed', linewidth=2)
plt.ylabel("Thsi is y axis")
plt.xlabel("This is x axis")

plt.subplot(1,3,2) # First arg represent the n-rows sec n-col thrid position of subplot
plt.title("This is first title")
plt.plot(x, y2, color='g', linestyle = 'dashed', linewidth=2)
plt.ylabel("Thsi is y axis")
plt.xlabel("This is x axis")

plt.grid(True)
plt.show()
```

```python
plt.subplot(3,1,1) # First arg represent the n-rows sec n-col thrid position of subplot
plt.plot(x, y1, color='m', linestyle = 'dotted', linewidth=2)

plt.subplot(3,1,3) # First arg represent the n-rows sec n-col thrid position of subplot
plt.plot(x, y2, color='b', linestyle = 'dashed', linewidth=2)

plt.subplot(3,1,2) # First arg represent the n-rows sec n-col thrid position of subplot
plt.plot(x, y2, color='g', linestyle = 'dashed', linewidth=2)

plt.grid(True)
plt.show()
```

```python
fig,a = plt.subplots(2,2)
a[0][0].set_title("This is first plot")
a[0][0].set_xlabel("This is x axis")
a[0][0].plot(x, y2, color='g', linestyle = 'dashed', linewidth=2)

a[0][0].set_title("This is first plot")
a[0][0].set_xlabel("This is x axis")
a[0][1].plot(x, y2, color='r', linestyle = 'dashed', linewidth=2)

a[0][0].set_title("This is first plot")
a[0][0].set_xlabel("This is x axis")
a[1][0].plot(x, y2, color='y', linestyle = 'dashed', linewidth=2)

a[0][0].set_title("This is first plot")
a[0][0].set_xlabel("This is x axis")
a[1][1].plot(x, y2, color='b', linestyle = 'dashed', linewidth=2)

plt.show()
```

# Bar plot

```python
students = {"Sumit":87,"Gupta":99,"Manohar":66,"Ajay":35}
names = list(students.keys())
marks = list(students.values())
```

```python
plt.bar(names,marks)
plt.show()
```

# Adding title and label and grid

```python
plt.bar(names,marks,color='g', width=0.4)
plt.title("This is bar graph")
plt.xlabel("Thsi is x axis")
plt.ylabel("This is y axis")
plt.grid()
plt.show()
```

# Horizontal bar plot

```python
plt.barh(names,marks,color='m')
plt.title("This is bar graph")
plt.xlabel("Thsi is x axis")
plt.ylabel("This is y axis")
plt.grid()
plt.show()
```

# Changing mark aesthistecs

```python
x = list(range(10,100,10)) # both should be same soze
a = [3,4,7,5,2,6,4,8,9]
plt.scatter(x, a, marker="*", color="g", s=100)
plt.show()
```

# Adding two markers in the same plot

```python
x = list(range(10,100,10)) # both should be same soze
a = [3,4,7,5,2,6,4,8,9]
b = [0,8,9,3,9,4,5,1,3]
plt.scatter(x, a, marker="*", color="r", s=100)
plt.scatter(x, b, marker=".", color="g", s=150)
plt.show()
```

# Adding sub-plots

```python
x = list(range(10,100,10)) # both should be same soze
a = [3,4,7,5,2,6,4,8,9]
b = [0,8,9,3,9,4,5,1,3]

plt.subplot(1,2,1)
plt.scatter(x, a, marker="*", color="r", s=100)

plt.subplot(1,2,2)
plt.scatter(x, b, marker=".", color="g", s=150)
plt.show()
```

# Pie Chart

```python
fruts = ["apple","mongo","babana","lamen"]
quantity = [55,26,10,69]
```

```python
plt.pie(quantity,labels=fruts)
plt.show()
```

# changing Aesthelics

```python
plt.pie(quantity,labels=fruts, autopct="%0.1f%%")
plt.show()
```

```python
plt.pie(quantity,labels=fruts, autopct="%0.1f%%",colors=["gray","blue","r","g"])
plt.show()
```

# DoughNut Chart

```python
plt.pie(quantity, labels=fruts,radius = 2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
